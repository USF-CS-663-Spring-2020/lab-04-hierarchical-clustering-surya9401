```python
import pandas as pd
df = pd.read_csv("NIPS_1987-2015.csv")
```

```python
# Dropping column since it has non-numeric values
df_new = df.drop(columns = 'Unnamed: 0')
df_new
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df_new, 'ward')
```

```python
#Plotting Dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(20,15))
dendrogram(Z)
plt.show()
```
