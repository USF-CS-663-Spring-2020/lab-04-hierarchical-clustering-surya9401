```python
import pandas as pd
df = pd.read_csv("USCensus1990-2000rows.data.txt")
df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
Z = linkage(df, 'ward')
```

```python
# Plotting Dendrogram 
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(20,10))
dendrogram(Z)
plt.show()
```
